##Using the pandas dummies

```python
dum = pd.get_dummies(df['Car Model']).astype(int)
dum
```

```python
merge = pd.concat([df,dum],axis='columns')
merge
```

```python
final = merge.drop(['Car Model'],axis='columns')
final
```

```python
final = final.drop(['Mercedez Benz C class'],axis='columns')
final
```

```python
x = final.drop(['Sell Price($)'],axis='columns')
x
```

```python
y = final['Sell Price($)']
y
```

```python
model = linear_model.LinearRegression()
model.fit(x,y)
```

```python
model.predict(x)
```

```python
model.score(x,y)
```

##One-hot encoding

```python
df = pd.read_csv('/content/drive/MyDrive/carprices.csv')
df
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('Car Model',OneHotEncoder(),[0])],remainder='passthrough')
ct
```

```python
X = df.drop(['Sell Price($)'],axis='columns')
X
```

```python
Y = df['Sell Price($)']
Y
```

```python
X = ct.fit_transform(X)
X
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X,Y)
```

```python
model.predict([[0,1,0,69000,5]])
```

```python
model.predict([[0,0,1,45000,4]])
```

```python
model.predict([[1,0,0,86000,7]])
```

```python
model.score(X,Y)
```
